```python
import os
import numpy as np
```

```python
str2='	symb yloc = 0.01655 * $coordFactor'
str3='	hist xyz aprs $xloc * * $yloc * *'
substr1='	symb xloc = '
substr2=' * $coordFactor'
str1=[]
for i in np.arange(0.0005,0.0395,0.00005):
    str1.append(substr1+str('%.5f'%i)+substr2)
print(str1)
file_content=[]
for i in range(len(str1)):
    file_content.append(str1[i])
    file_content.append(str2)
    file_content.append(str3)
print((file_content))
```

```python
with open('script.txt','w') as f:
    for item in file_content:
        f.write("%s\n" % item)
```
